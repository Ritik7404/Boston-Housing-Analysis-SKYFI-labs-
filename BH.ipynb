```python
import numpy as np 
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation

from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from IPython.display import HTML
```

# Step 1 : Data Preprocessing

```python
# Load dataset

boston = load_boston()

# Description

print(boston.DESCR)
```

```python
features = pd.DataFrame(boston.data,columns = boston.feature_names)
features
```

```python
target = pd.DataFrame(boston.target,columns=['target'])
target
```

```python
max(target['target'])
```

```python
min(target['target'])
```

```python
# Concatenate

df = pd.concat([features,target],axis=1)
df
```

```python
# round(decimals=2) to set the precision to 2 decimal palces

df.describe().round(decimals=2)
```

```python
# Calaculate correlation between every columns on the data
corr = df.corr(method='pearson')

# Take absolute values of corelation
corrs = [abs(corr[attr]['target']) for attr in list(features)]

# Make a list of pairs [(corr, features)]
l =list(zip(corrs, list(features)))

# Sort the list of pairs in reverse/descending order,
# With the corelation value as the key of sorting
l.sort(key = lambda x : x[0], reverse = True)

# Unzip pairs to 2 lists
corrs, labels = list(zip(*l))

# Plot corelation w.r.t target variables as a bar graph
index = np.arange(len(labels))
plt.figure(figsize=(15,5))
plt.bar(index,corrs,width=0.5)
plt.xlabel('Attributes')
plt.ylabel('Correlation with the target variables')
plt.xticks(index, labels)
plt.show()
```
